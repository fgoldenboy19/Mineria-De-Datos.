```python
import matplotlib.pyplot as plt
import pandas as pd

# Placeholder DataFrame - Replace this with your actual data loading
data = {
    'Horas_Estudio': [2, 5, 3, 6, 4, 7, 5, 8, 6, 9],
    'Calificacion_Final': [60, 75, 65, 80, 70, 85, 78, 90, 82, 95],
    'Asistencia': [8, 9, 7, 10, 8, 9, 9, 10, 7, 10],
    'Tareas': [7, 8, 6, 9, 7, 8, 8, 9, 7, 10],
    'Examen': [60, 70, 65, 80, 75, 85, 78, 90, 82, 95],
    'Edad': [18, 19, 18, 20, 19, 21, 20, 22, 21, 23],
    'Matricula': [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008, 1009, 1010],
    'Nombre': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi', 'Ivan', 'Judy']
}
df = pd.DataFrame(data)

#Practica 4

#Tabla de Dispercion Horas de estudio vs Calificacion Final

plt.scatter(df["Horas_Estudio"], df["Calificacion_Final"])
plt.xlabel("Horas de estudio")
plt.ylabel("Calificación Final")
plt.title("Horas de Estudio vs Calificación Final")
plt.show()

#Tabla de Dispercion Asistencia vs Calificacion Final

plt.scatter(df["Asistencia"], df["Calificacion_Final"])
plt.xlabel("Asistencia")
plt.ylabel("Calificación Final")
plt.title("Asistencia vs Calificación Final")
plt.show()

#Tabla de Dispercion Tareas vs Calificacion Final

plt.scatter(df["Tareas"], df["Calificacion_Final"])
plt.xlabel("Tareas")
plt.ylabel("Calificación Final")
plt.title("Tareas vs Calificación Final")
plt.show()

#Tabla de Dispercion Examen vs Calificacion Final

plt.scatter(df["Examen"], df["Calificacion_Final"])
plt.xlabel("Examen")
plt.ylabel("Calificación Final")
plt.title("Examen vs Calificación Final")
plt.show()

#Tabla de Dispercion Examen vs Calificacion Final

plt.scatter(df["Edad"], df["Calificacion_Final"])
plt.xlabel("Edad")
plt.ylabel("Calificación Final")
plt.title("Edad vs Calificación Final")
plt.show()

#Tabla de Dispercion Matricula vs Calificacion Final

plt.scatter(df["Matricula"], df["Calificacion_Final"])
plt.xlabel("Matricula")
plt.ylabel("Calificación Final")
plt.title("Matricula vs Calificación Final")
plt.show()



#Practica 4 (Tarea)

#Grafica de Barras (Asistencia)

plt.bar(df["Nombre"], df["Asistencia"])
plt.xlabel("Estudiante")
plt.ylabel("Asistencias")
plt.title("Asistencias por estudiante")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

#Histograma (Tareas)

plt.hist(df["Tareas"], bins=5)
plt.xlabel("Calificación de tareas")
plt.ylabel("Cantidad de estudiantes")
plt.title("Calificaciones de tareas")
plt.show()

#Boxplot (Examen)

plt.boxplot(df["Examen"])
plt.ylabel("Calificación del examen")
plt.title("Calificaciones del examen")
plt.show()
```
